This is a quick and hacky notebook to test how much overlap there is between the tokens in `output_ds.txt` and `output_rdd.txt` (which is pretty much identical to `output.txt`)

```python
import re

def extract_tokens(text):
    # Remove colons
    text = text.replace(":", "")
    # Remove periods
    text = text.replace(".", "")
    
    # Remove numbers using regular expression
    text = re.sub(r'\d+', '', text)

    # Remove text between angular brackets using regular expression
    text = re.sub(r'<.*?>', '', text)
    
    return text

# Example usage
text = "This is a string: with numbers 123 and colons: 45:67 and some <text> between angular brackets</text> and a dot."
processed_text = extract_tokens(text)
print(processed_text)
```

```python
def load_text_as_string(file):
    with open(file, 'r') as f:
        text = f.read()
    return text
```

```python
rdd = extract_tokens(load_text_as_string('output_rdd.txt'))
```

```python
rdd
```

```python
def separate_by_space_keep_if_not_empty(text):
    return [e for e in re.split(r'\s+', text) if e != '']
```

```python
separate_by_space_keep_if_not_empty(rdd)
```

```python
rdd_tokens = separate_by_space_keep_if_not_empty(rdd)
```

```python
ds = load_text_as_string('output_ds.txt')
```

```python
ds_tokens = separate_by_space_keep_if_not_empty(ds)
```

```python
ds_tokens = set(ds_tokens)
```

```python
rdd_tokens = set(rdd_tokens)
```

```python
def get_overlap_percentage(set1, set2):
    return len(set1.intersection(set2)) / len(set1)
```

```python
get_overlap_percentage(rdd_tokens, ds_tokens)
```

```python
get_overlap_percentage(ds_tokens, rdd_tokens)
```
